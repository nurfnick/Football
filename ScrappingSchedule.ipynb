```python
data_dict =[]
for table in tables:
    headers = []
    rows = table.find_all('tr')
    for header in table.find('tr').find_all('th'):
        headers.append(header.text)
    for row in table.find_all('tr')[1:]:
        values =[]
        for col in row.find_all(['th','td']):
            values.append(col.text)
        if values:
            episode_dict = {headers[i]: values[i] for i in range(len(values))}
            data_dict.append(episode_dict)  
```

```python
def tablecleaner(table_in_BeautifulSoup):
    data =[]
    for table in table_in_BeautifulSoup:
        rows = table.find_all('tr')
        for row in table.find_all('tr')[1:]:
            values =[]
            for col in row.find_all(['th','td']):
                values.append(col.text)
            if row.find('a') is None:
                values.append([])
            else:
                values.append(row.find('a').get('href'))
            data.append(values)
    return data
```

```python
masterfile =[]

for i in range (0,16):
    url = urlteamroster(teamdicturl[schedule2019[i,3]]) #away team
    r = requests.get(url)
    html_contents = r.text
    html_soup = BeautifulSoup(html_contents,"lxml")
    commented_table = html_soup.find('div', id ="all_games_played_team" ).find_all(string=lambda text: isinstance(text, Comment))
    tables = BeautifulSoup(commented_table[0], 'lxml')
    awaydata = tablecleaner(tables)
    url = urlteamroster(teamdicturl[schedule2019[i,6]]) #home team
    r = requests.get(url)
    html_contents = r.text
    html_soup = BeautifulSoup(html_contents,"lxml")
    commented_table = html_soup.find('div', id ="all_games_played_team" ).find_all(string=lambda text: isinstance(text, Comment))
    tables = BeautifulSoup(commented_table[0], 'lxml')
    homedata = tablecleaner(tables)
    masterfile.append([awaydata,homedata])
```
